```python
import pandas as pd
import matplotlib.pyplot as plt
```

# **Plot accelerometer and gyroscope data individually**

```python
def load_and_plot_data_individually(name):
  try:

      path = '/content/'
      csv_file_path = path + name + '.csv'
      png_file_path = path + name + '.png'
      output_image_path = path + name + '_output.png'

      df = pd.read_csv(csv_file_path) # Read the CSV file into a pandas DataFrame

      print(f"Successfully loaded data from '{csv_file_path}'.")
      print("\nFirst 5 rows of the data:")
      print(df.head())
      print("\nData information:")
      df.info()

      plt.figure(figsize=(12, 6))

      plt.plot(df['seconds_elapsed'], df['z'], label='Z-Axis', marker='.', linestyle='-')
      plt.plot(df['seconds_elapsed'], df['y'], label='Y-Axis', marker='.', linestyle='-')
      plt.plot(df['seconds_elapsed'], df['x'], label='X-Axis', marker='.', linestyle='-')

      plt.xlabel('Seconds Elapsed', fontsize=12)
      plt.ylabel('Value (Z, Y, X)', fontsize=12)
      plt.title('Time Series Plot of Z, Y, and X vs. Seconds Elapsed', fontsize=14)
      plt.grid(True, linestyle='--', alpha=0.7)
      plt.legend(loc='best', title='Variable')

      plt.tight_layout()
      plt.savefig(output_image_path)
      plt.show()

  except FileNotFoundError:
      print(f"Error: The file '{csv_file_path}' was not found. Please check the file path.")
  except KeyError as e:
      print(f"Error: A required column was not found. Please ensure your CSV has 'seconds_elapsed', 'z', 'y', and 'x' columns. Missing column: {e}")
  except Exception as e:
      print(f"An unexpected error occurred: {e}")
```

```python
load_and_plot_data_individually('Forehand Accelerometer')
```

```python
load_and_plot_data_individually('Forehand Gyroscope')
```

```python
load_and_plot_data_individually('Backhand Accelerometer')
```

```python
load_and_plot_data_individually('Backhand Gyroscope')
```
